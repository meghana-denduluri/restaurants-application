# Entity Resolution for Dishes-Recipes

```
dishes = pd.read_csv(path+'/Database_Data/Dishes.csv').iloc[:,1:]
recipes = pd.read_csv(path+'/Database_Data/Recipes.csv').iloc[:,1:]
```

```
from  py_stringmatching.tokenizer.qgram_tokenizer import QgramTokenizer
from py_stringsimjoin.join.jaccard_join import jaccard_join
```

```
tknzr = QgramTokenizer(qval=5,padding=True)
dishes["dishes"] = dishes.dishes.apply(lambda x: x.lower())
recipes["name"] = recipes.name.astype(str).apply(lambda x: x.lower())
```

```
#Threshold for jaccard similarity
thresh = .4

answer_df = jaccard_join(dishes, recipes, "index", "id",
                         "dishes", "name", tknzr, thresh,
                          l_out_attrs=["dishes"], 
                          r_out_attrs=["name"], n_jobs=-1, 
                          l_out_prefix="dishes_",r_out_prefix="recipes_",
                         show_progress=True)\
                          .drop_duplicates()
display(answer_df)
```

```
RecipeOf = answer_df[['dishes_index','recipes_id']].drop_duplicates()
RecipeOf
```
